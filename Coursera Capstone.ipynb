```python
# Getting data from Wikipedia
import pandas as pd
import requests
from bs4 import BeautifulSoup

req = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(req.content,'lxml')
table = soup.find_all('table')[0]
df = pd.read_html(str(table))
toronto_raw_data=pd.DataFrame(df[0])

toronto_raw_data
```

```python
# Eliminate rows where Borough is 'Not Assigned' by keeping rows where Borough is not 'Not assigned'
toronto_df = toronto_raw_data[toronto_raw_data['Borough'] != 'Not assigned'].reset_index(drop=True)
toronto_df
```
